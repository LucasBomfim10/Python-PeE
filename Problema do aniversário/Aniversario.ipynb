# Problema do Aniversário

Neste projeto, estamos analisando o problema do aniversário, que é um problema clássico em probabilidade. Vamos calcular a probabilidade de duas pessoas terem a mesma data de aniversário, e ver como essa probabilidade varia com o número de pessoas.

Primeiro, vamos importar as bibliotecas necessárias:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Agora, vamos definir uma função para calcular a probabilidade de duas pessoas terem a mesma data de aniversário:

```python
def Probabilidade_Aniversario(Num_Pessoas):
    Dias = 365
    Probabilidade = 1
    for i in range(Num_Pessoas):
        Probabilidade = np.prod((Dias - np.arange(Num_Pessoas)) / Dias)
    return ((1 - Probabilidade)*100)
```

Vamos testar com um número de pessoas especifico, execute o codigo e digite um número

```python
while True:
    Num_Pessoas = input("Digite o número de pessoas: ")
    if Num_Pessoas.isdigit() and int(Num_Pessoas) > 0 :
        Num_Pessoas = int(Num_Pessoas)
        break
    else:
        print("Digite um número inteiro e maior que 0")
```

Conferindo o resultado

```python
print(f"A probabilidade de duas pessoas terem a mesma data de aniversário é de: {Probabilidade_Aniversario(Num_Pessoas):.4f}%")
```

Agora vamos analisar como a probabilidade de duas pessoas terem a mesma data de aniversário varia com o número de pessoas no total

Primeiro criamos uma lista "Probabilidades" e adicionamos o resultado de cada operaçao para as diferentes quantidades de pessoas

```python
Probabilidades = []

for i in range(1,366):
    Probabilidades.append(Probabilidade_Aniversario(i))
```

Usamos o matplotlib para ver o resultado

```python
plt.figure()

plt.plot(range(1,366), Probabilidades)

plt.title("Probabilidade por número de pessoas")

plt.xlabel("Número de pessoas")
plt.ylabel("Probabilidade (%)")

plt.show()
```

Passamos o resultado para um dataframe do pandas e em seguida exportamos para csv para utilizar com outra ferramenta como EXCEL

```python
df = pd.DataFrame({"Número de pessoas": range(1,366), "Probabilidade (%)": Probabilidades})

df.to_csv("Probabilidade_Aniversario.csv", index = False)
```
